## Exploratory Data Analysis

### Check for Missing Data

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='plasma') 
# we can see  that there is no missing data in the dataset we can proceed furthur.Note: yellow lines on the blue space represent the missing data.
```

```python
sns.set_style('darkgrid')
sns.countplot(x="presence of Heart Disease",data=df)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='presence of Heart Disease',hue='Sex',data=df)# 0:female,1:male
```

```python
df['age'].hist(bins=20)
```

```python
df['resting blood pressure'].hist(color='green',bins=30)
```

```python
corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```
